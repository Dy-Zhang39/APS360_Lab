### Part (a) Convolutional Network - 5 pt

Build a convolutional neural network model that takes the (224x224 RGB) image as input, and predicts the gesture
letter. Your model should be a subclass of nn.Module. Explain your choice of neural network architecture: how
many layers did you choose? What types of layers did you use? Were they fully-connected or convolutional?
What about other decisions like pooling layers, activation functions, number of channels / hidden units?

```python
class GestureClassifier(nn.Module):
      def __init__(self):
          self.name = "GestureClassifier"
          super(GestureClassifier, self).__init__()
          self.conv1 = nn.Conv2d(3, 5, 5) #in_channels = 3, out_channels = 5, kernel_size = 5, output size: (224-5)/1+1=220
          self.pool = nn.MaxPool2d(2, 2) #summary the feature, output size: (220-2)/2+1=110
          self.conv2 = nn.Conv2d(5, 10, 5) #in_channels, out_chanels, kernel_size, output size:(110-5)/1+1=106
          self.fc1 = nn.Linear(10*53*53, 32) #after maxpool: output size: (106-2)/2+1=53, channel: 10
          self.fc2 = nn.Linear(32, 9) #total 9 classes

      def forward(self, x):
          x = self.pool(F.relu(self.conv1(x)))  #con1 and maxpool
          x = self.pool(F.relu(self.conv2(x)))  #con2 and maxpool
          x = x.view(-1, 10*53*53)              #flatten    
          x = F.relu(self.fc1(x))               #fully connected layer
          x = self.fc2(x)                       #classify
          x = x.squeeze(1)
          return x
```

A CNN model with 2 convolutional layers, 2 maxpooling layers, and 2 fully connected layers are chosen. The input to fisrt convolutional layer is 3 RGB input channels, with out channels and kernel size set to 5. This layer extracts local features of the images. A max pool with fiter size 2*2 and stride 2 is applied to summarize the features. The second convolutional layer takes 5 input channels, 10 output channels, and kernel size of 5, this layer extracts more abstract features of the images, a max pool is also applied to sumarize the feature. The activation function is chosen to be ReLU, which introduce some non-linearities for our model. Finaaly, two fully connected layers are applied at the end to classify the images.

### Part (b) Training Code - 5 pt

Write code that trains your neural network given some training data. Your training code should make it easy

```python
import torch.optim as optim
```

```python
def get_model_name(name, batch_size, learning_rate, epoch):
    """ Generate a name for the model consisting of all the hyperparameter values

    Args:
        config: Configuration object containing the hyperparameters
    Returns:
        path: A string with the hyperparameter name and value concatenated
    """
    path = "model_{0}_bs{1}_lr{2}_epoch{3}".format(name,
                                                   batch_size,
                                                   learning_rate,
                                                   epoch)
    return path
```

```python
def get_accuracy(model, loader):

    data_loader = loader

    criterion = nn.CrossEntropyLoss()
    correct = 0
    total = 0
    total_loss = 0
    sample_size = 0
    for imgs, labels in data_loader:

        #############################################
        #To Enable GPU Usage
        if use_cuda and torch.cuda.is_available():
          imgs = imgs.cuda()
          labels = labels.cuda()
        #############################################

        output = model(imgs)
        loss = criterion(output, labels)
        total_loss += loss
        
        #select index with maximum prediction score
        pred = output.max(1, keepdim=True)[1]
        correct += pred.eq(labels.view_as(pred)).sum().item()
        total += imgs.shape[0]
        sample_size += 1
    return correct / total , total_loss / sample_size
```

```python
def train(model, train_loader, val_loader, batch_size=20, num_epochs=30, alpha=0.01, num_workers=1):
    #train_loader = torch.utils.data.DataLoader(train_data, batch_size=batch_size, 
    #                                       num_workers=num_workers, shuffle=True)
    #val_loader = torch.utils.data.DataLoader(val_data, batch_size=batch_size, 
    #                                   num_workers=num_workers, shuffle=True)

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=alpha)

    iters, train_loss, train_acc, val_acc, val_loss = [], [], [], [], []

    # training
    n = 0 # the number of iterations
    print("start training...")
    start_time=time.time()
    for epoch in range(num_epochs):

        for imgs, labels in iter(train_loader):
          
            
            #############################################
            #To Enable GPU Usage
            if use_cuda and torch.cuda.is_available():
              imgs = imgs.cuda()
              labels = labels.cuda()
            #############################################

            optimizer.zero_grad()         # a clean up step for PyTorch
            out = model(imgs)             # forward pass
            loss = criterion(out, labels) # compute the total loss
            loss.backward()               # backward pass (compute parameter updates)
            optimizer.step()              # make the updates for each parameter
            
            # save the current training information
            iters.append(n)
            trainning_ac, trainining_loss = get_accuracy(model, train_loader)
            train_loss.append(float(loss)/batch_size)             # compute *average* loss
            train_acc.append(float(trainning_ac))
            val_ac, val_lo = get_accuracy(model, val_loader)
            val_acc.append(float(val_ac))  # compute validation accuracy
            val_loss.append(float(val_lo)/batch_size)

            n += 1
            print("iteration: ",n,'Progress: % 6.2f ' % ((epoch * len(train_loader)) / (num_epochs * len(train_loader))*100),'%', "Time Elapsed: % 6.2f s " % (time.time()-start_time))
            
            print("train acc is: ",train_acc[-1], "Val acc: ", val_acc[-1])
            # Save the current model (checkpoint) to a file
            model_path = get_model_name(model.name, batch_size, alpha, epoch)
            torch.save(model.state_dict(), model_path)

        print ("Epoch %d Finished. " % epoch ,"Time per Epoch: % 6.2f s "% ((time.time()-start_time) / (epoch +1)))

    end_time= time.time()# training
    print("#######################traning finished#################################")
    
    # plotting
    plt.title("Training Curve")
    plt.plot(iters, train_loss, label="Train")
    plt.xlabel("Iterations")
    plt.ylabel("Loss")
    plt.show()

    plt.title("Training Curve")
    plt.plot(iters, train_acc, label="Training")
    plt.plot(iters, val_acc, label="Validation")  
    plt.xlabel("Iterations")
    plt.ylabel("Validation Accuracy")
    plt.legend(loc='best')
    plt.show()
    
    print ("Total time:  % 6.2f s  Time per Epoch: % 6.2f s " % ( (end_time-start_time), ((end_time-start_time) / num_epochs) ))

    return train_loss, val_loss, train_acc, val_acc
```

### 4. Transfer Learning [15 pt]
For many image classification tasks, it is generally not a good idea to train a very large deep neural network
model from scratch due to the enormous compute requirements and lack of sufficient amounts of training
data.

One of the better options is to try using an existing model that performs a similar task to the one you need
to solve. This method of utilizing a pre-trained network for other similar tasks is broadly termed **Transfer
Learning**. In this assignment, we will use Transfer Learning to extract features from the hand gesture
images. Then, train a smaller network to use these features as input and classify the hand gestures.

As you have learned from the CNN lecture, convolution layers extract various features from the images which
get utilized by the fully connected layers for correct classification. AlexNet architecture played a pivotal

Build a convolutional neural network model that takes as input these AlexNet features, and makes a

```python
class AlexNetCNN(nn.Module):
    def __init__(self):
        super(AlexNetCNN, self).__init__()
        self.name = "AlexNetCNN"
        self.conv1 = nn.Conv2d(256, 32, 3)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(800, 32)
        self.fc2 = nn.Linear(32, 9)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = x.view(-1, 800)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        x = x.squeeze(1) # Flatten to [batch_size]
        return x 
```

```python
def train_AlexNet(model, AlexNet_train, AlexNet_train_label, AlexNet_val, AlexNet_val_label, batch_size=200, num_epochs=30, alpha=0.01, num_workers=1):
    train_loader = torch.utils.data.DataLoader(AlexNet_train, batch_size=batch_size, 
                                           num_workers=num_workers, shuffle=True)
    val_loader = torch.utils.data.DataLoader(AlexNet_val, batch_size=batch_size, 
                                       num_workers=num_workers, shuffle=True)

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=alpha)

    iters, train_loss, train_acc, val_acc, val_loss = [], [], [], [], []

    # training
    n = 0 # the number of iterations
    start_time=time.time()
    for epoch in range(num_epochs):

            imgs, labels = AlexNet_train, AlexNet_train_label
          
            
            #############################################
            #To Enable GPU Usage
            if use_cuda and torch.cuda.is_available():
              imgs = imgs.cuda()
              labels = labels.cuda()
            #############################################
          
            out = model(imgs)             # forward pass
            loss = criterion(out, labels) # compute the total loss
            loss.backward()               # backward pass (compute parameter updates)
            optimizer.step()              # make the updates for each parameter
            optimizer.zero_grad()         # a clean up step for PyTorch

            trainning_ac, trainining_loss = get_accuracy(model, AlexNet_train, AlexNet_train_label)

            # save the current training information
            iters.append(n)
            train_loss.append(float(loss)/batch_size)             # compute *average* loss
            train_acc.append(float(trainning_ac))
            val_ac, val_lo = get_accuracy(model, AlexNet_val, AlexNet_val_label)
            val_acc.append(float(val_ac))  # compute validation accuracy
            val_loss.append(float(val_lo)/batch_size)

            n += 1
            print("Iteration: ",n,'Progress: % 6.2f ' % ((epoch * len(train_loader)) / (num_epochs * len(train_loader))*100),'%', "Time Elapsed: % 6.2f s " % (time.time()-start_time))
            
            print("train acc is: ",train_acc[-1])
            # Save the current model (checkpoint) to a file
            model_path = get_model_name(model.name, batch_size, alpha, epoch)
            torch.save(model.state_dict(), model_path)

            print ("Epoch %d Finished. " % epoch ,"Time per Epoch: % 6.2f s "% ((time.time()-start_time) / (epoch +1)))

    end_time= time.time()# training
    print("#######################traning finished#################################")
    
    # plotting
    plt.title("Training Curve")
    plt.plot(iters, train_loss, label="Train")
    plt.xlabel("Iterations")
    plt.ylabel("Loss")
    plt.show()

    plt.title("Training Curve")
    plt.plot(iters, train_acc, label="Training")
    plt.plot(iters, val_acc, label="Validation")  
    plt.xlabel("Iterations")
    plt.ylabel("Validation Accuracy")
    plt.legend(loc='best')
    plt.show()
  
    print ("Total time:  % 6.2f s  Time per Epoch: % 6.2f s " % ( (end_time-start_time), ((end_time-start_time) / num_epochs) ))

    return train_loss, val_loss, train_acc, val_acc
```
